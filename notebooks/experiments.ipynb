```python
import os
import torch
import librosa
import soundfile as sf
import matplotlib.pyplot as plt
from IPython.display import Audio, display
```

```python
AUDIO_PATH = "oiseaux.wav"  # mets ton propre fichier ici
TARGET_SR = 44100 # Sampling rate (Hz) 
DURATION = 20  # seconds

# Load audio
x, sr = librosa.load(AUDIO_PATH, sr=TARGET_SR, mono=True)
x = x[:TARGET_SR * DURATION]

print(f"Loaded audio: {x.shape}, sr={sr}")

display(Audio(x, rate=sr))
```

```python
FILE_NAME = "oiseaux.wav"
TRAIN_WAV = f"../Catch-A-Waveform/inputs/{FILE_NAME}"
sf.write(TRAIN_WAV, x, TARGET_SR)
print("Training audio saved:", TRAIN_WAV)
```
